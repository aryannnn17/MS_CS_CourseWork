```python
# Name: Aryan Jigneshbhai Bhagat
# NetID: sl5310
# Email: abhagat4@horizon.csueastbay.edu

# ------------< Coding Assignment: K-Nearest Neighbor >-----------------------------

# Question:
# Using the given dataset below (diabetes.csv), write a code, preferably in Jupyter Notebook, that uses kNN classifier to classify the new data. For this problem, you are not to use libraries for the kNN. Submit the Jupyter Notebook file. Make sure the program is your own work.

# Solution:
# -------------------Modules-----------------------
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

#------------------Data file--------------------------
data = pd.read_csv('diabetes.csv')

print("Dataset Details:")
print(data.head())

print("\nDataset Info:")
print(data.info())

x = data.iloc[:, :-1].values 
y = data.iloc[:, -1].values  

# --------------------Train Model----------------------------
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)

xTrain, xTest, yTrain, yTest = train_test_split(x_scaled, y, test_size=0.2, random_state=42)

def accScore(y_true, y_pred):
    corr = sum(y_true == y_pred)
    return corr / len(y_true) * 100

def classifyTest(xTrain, yTrain, xTest, k=3):
    predictions = []
    for test_instance in xTest:
        prediction = kNN(xTrain, yTrain, test_instance, k=k)
        predictions.append(prediction)
    return predictions

def euclid_dist(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))

def kNN(xTrain, yTrain, xTest_instance, k=3):
    dist = []
    for i in range(len(xTrain)):
        distances = euclid_dist(xTest_instance, xTrain[i])
        dist.append((distances, yTrain[i]))

    dist.sort(key=lambda x: x[0])

    neighbors = dist[:k]

    class_votes = {}
    for _, label in neighbors:
        if label in class_votes:
            class_votes[label] += 1
        else:
            class_votes[label] = 1

    predicted_class = max(class_votes, key=class_votes.get)
    return predicted_class


# ------------------------Test Model---------------------------
k = 5  
y_pred = classifyTest(xTrain, yTrain, xTest, k=k)
accuracy = accScore(yTest, y_pred)
newData = np.array([[0.5, 0.2, 0.6, 0.3, 0.4, 0.5, 0.6, 0.7]]) 
newPred = kNN(xTrain, yTrain, newData[0], k=k)

print(f"\n\nOutput: ")
print(f"Accuracy of the custom kNN classifier: {accuracy:.2f}%")
print(f"Classification {newData}: {newPred}")
```
